## Mutex

```python
import random
random.seed(18)

def run_mutex(x_df, pairs, permutations =2000, saving_file='mutex' ):
    results=[]
#     df = x_df.copy()
#     df = df.T
    df=x_df.astype(int)
    m = MutEx(background=df, permutations=permutations)
    for g1, g2 in pairs:
        print (g1)
        print (g2)
        r = m.calculate([g1, g2], parallel=True)
        print (r.mutex_pvalue, r.co_occurence_pvalue)
        results.append(r)
    
    results_dict_list=[]
    for r in results:
        g1, g2 = r.items
        results_dict={}
        results_dict['g1'] = g1
        results_dict['g2'] = g2
        results_dict['mutex_pvalue'] = r.mutex_pvalue
        results_dict['co_occurence_pvalue'] = r.co_occurence_pvalue
        results_dict['zscore'] = r.zscore
        results_dict['signal'] = r.signal
        results_dict['coverage'] = r.coverage
        results_dict['signal_coverage_ratio'] = r.signal_coverage_ratio
        results_dict_list.append(results_dict)
    results_df = pd.DataFrame(results_dict_list)
    cols= ['g1', 'g2', 'mutex_pvalue', 'co_occurence_pvalue', 'zscore', 'signal', 'coverage', 'signal_coverage_ratio']
    results_df = results_df.reindex(cols, axis=1)
    results_df.to_csv('{}.csv'.format(saving_file))
    return results_df
```

## cBio (fisher exact)

```python
from scipy.stats import chi2_contingency
import scipy.stats as stats
from sklearn import metrics
import numpy as np

def compare_pair(df, g1, g2):
    cm = metrics.confusion_matrix(df[g1], df[g2], labels=[0,1])
    print(cm)
    A= cm[1,1] #both
    B= cm[1,0] # G1 not G2
    C= cm[0,1] # G2 not G1
    D= cm[0,0] #neither
    OR= (A*D)/(B*C)
    log2_OR = np.log2(OR)
    results_dict={}
    results_dict['G1'] = g1
    results_dict['G2'] = g2
    results_dict['both'] = A
    results_dict['G1 not G2'] = B
    results_dict['G2 not G1'] = C
    results_dict['neither'] = D
    results_dict['OR'] = OR
    results_dict['log2_OR'] = log2_OR
    
    chi2, chi2_pvalue, dof, ex = chi2_contingency(cm)
    results_dict['p_value_chi2'] = chi2_pvalue
    oddsratio, p_value= stats.fisher_exact(cm)
    results_dict['p_value_fisher'] = p_value
#     results_dict['oddsratio'] = oddsratio
    return results_dict

def run_cbio(df, pairs, saving_file):
    results_list =[] 
    for p in pairs:
        g1, g2 =p
        print (g1, g2)
        results_dict = compare_pair(df.T, g1, g2) 
        results_list.append(results_dict)
#         print (results_dict)
    results_df = pd.DataFrame(results_list)
    def label_race (row):
        if row['OR'] >=1.  :
            return 'Co-occurrence'
        else:
            return 'Mutual exclusivity'
    
    results_df['Tendency'] =results_df.apply (lambda row: label_race(row), axis=1)

    results_df ['OR'] = results_df['OR'].map('{:,.3f}'.format)
    results_df ['log2_OR'] = results_df['log2_OR'].map('{:,.3f}'.format)
    results_df ['p_value_fisher'] = results_df['p_value_fisher'].map('{:,.3f}'.format)
    results_df ['p_value_chi2'] = results_df['p_value_chi2'].map('{:,.3f}'.format)
    
    cols= ['G1','G2', 'neither' , 'G1 not G2', 'G2 not G1', 'both','OR',  'log2_OR', 'p_value_chi2', 'p_value_fisher','Tendency']
    results_df = results_df.reindex(cols, axis=1)
    results_df.to_csv('{}.csv'.format(saving_file))
    return results_df

```
