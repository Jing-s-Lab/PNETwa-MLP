```python
mapping_file = join(PROSTATE_DATA_PATH, 'raw_data/sample_mapping.tsv')
```

```python
mapping_id = pd.read_csv(mapping_file, sep='\t')
```

```python
mapping_id = mapping_id[['Tumor_Sample_Barcode','tpm_col']]
```

```python
mapping_id.dropna(subset = ['tpm_col'], inplace=True)
```

```python
mapping_id = mapping_id.set_index('tpm_col')
```

```python
mapping_id.shape
```

```python
df  = df.join(mapping_id, how='inner', on = 'tpm_id')
```

```python
df.head()
```

```python
read_counts = pd.pivot_table(df,index="HUGO_gene",columns=["Tumor_Sample_Barcode"],values="expected_count")
```

```python
read_counts.head()
```

```python
read_counts = read_counts.astype('int32')
```

```python
read_counts.head()
```

## get interesting gene amplifications

```python
filename = join(PROSTATE_DATA_PATH,'processed/P1000_data_CNA_paper.csv')

cnv = pd.read_csv(filename, index_col=0)
```

```python
cnv.head()
```

```python
new_index = read_counts.columns
```

```python
cnv_updated = cnv.reindex(new_index)
```

```python
cnv_updated.head()
```

```python
cnv_MDM4 = cnv_updated[['MDM4']].copy()
```

```python
cnv_MDM4.head()
```

```python
mapp_dict= {-2:'Low', -1:'Low', 0:'Low', 1.0:'High', 2.0: 'High'}
```

```python
mapp_dict
```

```python
cnv_MDM4.MDM4.replace(mapp_dict, inplace=True)
```

```python
cnv_MDM4.head()
```

```python
response_file = join(PROSTATE_DATA_PATH,'processed/response_paper.csv')
response = pd.read_csv(response_file, index_col =0)
response.head()
```

```python
filename = join(PROSTATE_DATA_PATH,'processed/MDM4_amps.csv')
to_be_Saved = cnv_MDM4.join(response)
to_be_Saved.to_csv(filename)
```

```python

cnv_PDGFA = cnv_updated[['PDGFA']].copy()
cnv_NOTCH1 = cnv_updated[['NOTCH1']].copy()
```

```python
cnv_NOTCH1.NOTCH1.replace(mapp_dict, inplace=True)
cnv_PDGFA.PDGFA.replace(mapp_dict, inplace=True)
```

```python
filename = join(PROSTATE_DATA_PATH,'processed/NOTCH1_amps.csv')
to_be_Saved = cnv_NOTCH1.join(response)
to_be_Saved.to_csv(filename)



filename = join(PROSTATE_DATA_PATH,'processed/PDGFA_amps.csv')
to_be_Saved = cnv_PDGFA.join(response)
to_be_Saved.to_csv(filename)

```
